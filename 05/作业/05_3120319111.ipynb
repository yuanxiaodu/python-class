```python
y=np.random.normal(loc=0.5,scale=0.4,size=1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))
```

```python
plt.figure()
```

```python
plt.subplot(221)
plt.plot(x,y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
```

```python
plt.subplot(222)
plt.plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
```

```python
plt.subplot(223)
plt.plot(x,y-y.mean())
plt.yscale('symlog',linthresh=0.01)
plt.title('symlog')
plt.grid(True)
```

```python
plt.subplot(224)
plt.plot(x,y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
plt.subplots_adjust(top=0.98,bottom=0.02,left=0.1,right=0.96,hspace=0.22,
                   wspace=0.25)
plt.show()
```
