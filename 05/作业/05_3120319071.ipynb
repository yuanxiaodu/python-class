```python
bikes = ['trek','cannonable','redline','specialized']  #列表
print(bikes)
#输出['trek', 'cannonable', 'redline', 'specialized'],包括了方括号
#访问列表中的元素
print(bikes[0])  #trek,索引从0开始
print(bikes[0].title())  #Trek，开头字母大写
print(bikes[-1])  #specialized,列表里最后一个，倒数第二个为-2
message = f"My first bike was a {bikes[0].title()}"
message #'My first bike was a Trek'
print(message)  #My first bike was a Trek
##print输出无引号
###############3.2修改、添加和删除元素
motorcycles = ['honda','yamaha','suzuki']
#修改元素
print(motorcycles) #['honda', 'yamaha', 'suzuki']
motorcycles[0] = 'ducati'
print(motorcycles) #['ducati', 'yamaha', 'suzuki'],第一个元素被修改
#末尾加元素(append())
motorcycles.append('ducati') #附加append
print(motorcycles) #['honda', 'yamaha', 'suzuki', 'ducati'],在末尾加元素
#中间插入元素(insert())
motorcycles.insert(0,'ducati') #知道新元素的索引和值
print(motorcycles) #['ducati', 'honda', 'yamaha', 'suzuki']
#删除元素
#用del删除——索引
del motorcycles[0] #删除第一个
print(motorcycles) #['yamaha', 'suzuki']
#用pop()删除——末尾
popped_motorcycle = motorcycles.pop() #弹出pop，删除列表末尾元素
print(motorcycles)  #['honda', 'yamaha'],核实是否删除了一个值
print(popped_motorcycle)  #suzuki,访问被删除的值
last_owned = motorcycles.pop()
print(f"The last motorcycle I owned was a {last_owned.title()}")
#The last motorcycle I owned was a Suzuki
#pop()用法：if列表中摩托车按购买时间储存，可指出最后购买的是哪款
#弹出列表中任意位置的元素——索引
first_owned = motorcycles.pop(0)
print(f"The first motorcycle I owned was a {first_owned.title()}")
#The first motorcycle I owned was a Honda
#每次用pop,被弹出的元素就不再列表内了
#根据值删除元素(remove())——值
motorcycles = ['honda', 'yamaha', 'suzuki','ducati'] #定义列表
motorcycles.remove('ducati') #删除变量ducati
too_expensive = 'ducati' #将'ducati'赋给变量too_expensive
print(f"\nA {too_expensive.title()} is too expensive for me")
#
#A Ducati is too expensive for me
##方法remove()只能删除第一个指定的值，如果要删除的值在列表中出现多次，需要用循环来实现
#############3.3组织列表
cars = ['bmw','audi','toyota','subaru']
#用sort()对列表永久排序
cars.sort() #按字母顺序排序
print(cars)  #['audi', 'bmw', 'subaru', 'toyota']
cars.sort(reverse=True)  #与字母顺序相反排列
print(cars)  #['toyota', 'subaru', 'bmw', 'audi']
#用sorted()对列表临时排序
```
