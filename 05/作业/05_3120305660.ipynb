### 1.2 The object-oriented interface and the pyplot interface
- OO-style
- pyplot-style

```python
# OO-style
x = np.linspace(-1, 1.5, 50)
#从哪画到哪第三个参量是画几个点,linspace类似range
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='Linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x\'s label')  # Add an x-label to the axes.
ax.set_ylabel('y\'s label')  # Add a y-label to the axes.
ax.set_title("A Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
# pyplot-style

x = np.linspace(0, 2, 50)
#一个是set xlable 另一个是plt。xlabel
plt.plot(x, x, label='Linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='Quadratic')  # etc.
plt.plot(x, x**3, label='Cubic')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("A Simple Plot")
plt.legend()
```

### 1.4 Plotting with keyword strings

```python
# Fixing random state for reproducibility
np.random.seed(19930410)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0., scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
#yan mo
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(figsize=(20,20))

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
#y轴的尺度表示
plt.title('log')
plt.grid(False)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
#.meanqu pingjun
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
#wo yao hua224zhege tu le
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

## 2 Overview of seaborn plotting functions
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
### 2.1 Similar functions for similar tasks

```python
import seaborn as sns
import pandas as pd
sns.set_theme(style="darkgrid")
# sns.set_theme(style=)
# Samuel Norman Seaborn is a fictional character portrayed 
# by Rob Lowe on the television serial drama The West Wing. 
#读入三个csv格式数据
penguins = pd.read_csv("penguins.csv")
tips = pd.read_csv('tips.csv')
fmri = pd.read_csv('fmri.csv')
```

```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack",bins=30)
#histogram 直方图的意思 , x=后边的参数应该是数据集中的一栏 hue后边也是 bin表示有多少条
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="fill")
#kde kernel density estimation核密度估计图，非参数检验的一种方法
#multiple是 显示子类的方式的参数 multiple{“layer”, “dodge”, “stack”, “fill”}有4个可选值 dodge会报错
#hue 代表的是子类别
```

### 2.2 Figure-level vs. axes-level functions


```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="fill")
#displot是图级别的画图
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="fill", kind="hist")
#kind使用kde代表是核函数形式 kind{“hist”, “kde”, “ecdf”}, 默认是hist
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species",kind='kde')
#col是以什么为标准，把图分开画的意思
```

```python
# Axes-level functions make self-contained plots¶
import matplotlib.pyplot as plt
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))#
#matplotlib.pyplot 会返回两个对象figure和axes构成的元组
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[1])
#ax指定画到那个子图上
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[0])
f.tight_layout()
#轴级别的自适应画图
#这个代码要加上import matplotlib.pyplot as plt
#
```

### 2.3 Visualizing statistical relationships

```python
# sns.relplot(x="total_bill", y="tip", data=tips); 
#x,y指定画图的数据列，data指定画图的数据
# sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips); 
#hue代表区分什么来画图
# sns.relplot(x="total_bill", y="tip", style="smoker", data=tips);
#相比上个图多了样式的区别，样式根据smoker来区分
# sns.relplot(x="total_bill", y="tip", hue="time", style="time", data=tips);
#相比之前把style换成了别的行列
# sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
# sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-0.2,l=.75", data=tips);
#palette基本是改变图示的步长
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
#size参数可以控制显示的大小
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
#sizes控制了图示的变化范围
sns.relplot(x="total_bill", y="tip", hue="time", col="day", data=tips);
#
```

```python
# More complex datasets will have multiple measurements for the same value of the x variable. 
# The default behavior in seaborn is to aggregate the multiple measurements at each x value by 
# plotting the mean and the 95% confidence interval around the mean:
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
# sns.relplot(x='timepoint',y="signal",kind='line',data=fmri)
# The confidence intervals are computed using bootstrapping, which can be time-intensive for 
# larger datasets. It’s therefore possible to disable them:
#ci参数是置信区间的英文简写
# sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);

# Another good option, especially with larger data, is to represent the spread of the 
# distribution at each timepoint by plotting the standard deviation instead of a confidence 
# interval:
#sd是分布范围的英文简写 spread of the distribution
# sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);

# To turn off aggregation altogether, set the estimator parameter to None This might produce 
# a strange effect when the data have multiple observations at each point.
#estimator是拟合器？ 
# sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
#hue是区分种类画同一个图上的参数
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);

# sns.relplot(x="timepoint", y="signal", hue="region", style="event",
#             dashes=False, markers=True, kind="line", data=fmri);
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=True,kind="line", data=fmri);
#dashes是用不用虚线，maker是用不用数据加点

sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'"))
#unit是把每个个体画了出来

sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri)
```
