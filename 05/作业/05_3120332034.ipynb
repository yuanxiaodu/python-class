```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
mu,sigma=100,15
x= mu+sigma*np.random.randn(10000)

n,bins,patches=plt.hist(x,50,density=1,facecolor='g',alpha=0.75)

plt.xlabel('Samrts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu=100,\\sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

```python
import re
```

```python
pattern = re.compile(r'^\w+([-+.]\w+)*@\w+([-.]\w+)*\.\w+([-.]\w+)*$')
def is_mail(email):
    if pattern.match(email):
        print("Email",email)
    else:
        print("Email:",email)
is_mail("01xian_+jiaoda@163.com")
is_mail("01xianjiaoda@")
is_mail("xian.jiaoda@baidu.com")
is_mail("www.baidu.com")

#18
pattern = re.compile(r'^((\d{18})|([0-9x]{18})|([0-9X]{18}))$')
def is_idcard(idcard):
    if pattern.match(idcard):
        print("",idcard)
    else:
        print("",idcard)

is_idcard("411624200502091815")
is_idcard("41162420050-181815")
is_idcard("411624nfb8155566xx")
```

```python
# 
class Dog:
    def __init__(self,name):
        self.__name = name
        self.__age = None
        print(self.__name,"")
    
    def set_age(self,age):
        if not isinstance(age,int):
            print("!")
            return False
        if age <=0:
            print("0!")
            return False
        self.__age = age
       

    def play(self):
        print(",",self.__age,"")
        
dog = Dog("")
dog.set_age("hello")
dog.set_age(-5)
dog.set_age(4)
dog.play()

# 
class Animal:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def play(self):
        print("%s"%self.name)
        print("%d"%self.age)

class Dog(Animal):
    def __init__(self):
        super(Dog,self).__init__("",3)
class Cat(Animal):
    pass
class Pig(Animal):
    pass
dog = Dog()
cat =Cat("",2)
pig =Pig("",1)
dog.play()
cat.play()
pig.play()
```
