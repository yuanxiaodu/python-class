```python
def flat(l):
    f = []
    for x in l:
        if type(x) is list:
            f.extend(x)    # 
        else:
            f.append(x)
    return f
flat(basket)
```

```python
def fibonacii(n):
    a, b, counter = 0, 1, 0
    while True:
        if (counter > n):
            return
        yield a
        a, b = b, a + b
        counter += 1
f = fibonacii(1)
for x in f:
    print(x, end=" ")
```

```python
class Parent(object):
    def override(self):
        print("PARENT override()")

    def implicit(self):
        print("PARENT implicit()")

    def altered(self):
        print("PARENT altered()")

class Child(Parent):
    def override(self):
        print("CHILD override()")

    def altered(self):
        print("CHILD, BEFORE PARENT altered()")
        super(Child, self).altered()
        print("CHILD, AFTER PARENT altered()")

dad = Parent()
son = Child()

dad.implicit()
son.implicit()
print('-' * 50)

dad.override()
son.override()
print('??'.center(50, '-'))

dad.altered()
son.altered()
```
