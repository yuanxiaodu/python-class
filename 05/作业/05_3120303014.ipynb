```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='一次')  # Plot some data on the axes.
ax.plot(x, x**2, label='二次')  # Plot more data on the axes...
ax.plot(x, x**3, label='三次')  # ... and some more.
ax.set_xlabel('x轴')  # Add an x-label to the axes.
ax.set_ylabel('y轴')  # Add a y-label to the axes.
ax.set_title("简单图形")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(221)
plt.plot(t1, f(-t1), 'bo', t2, f(t2), 'k')

plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.figure()

plt.subplot(223)
plt.plot(t1, np.sin(2*np.pi*t1), 'bo', t2, f(t2), 'k')

plt.subplot(224)
plt.plot(t2, np.tan(2*np.pi*t2), 'r--')
plt.suptitle('compare')
plt.show()
```
