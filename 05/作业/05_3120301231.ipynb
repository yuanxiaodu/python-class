```python
#面向对象
x = np.linspace(0,2,100) #起点，终点，等分个数
fig , ax = plt.subplots() #生成一个图像，和一个轴 fig设置图的size
ax.plot(x , x , label = 'linear')
ax.plot(x , x**2 , label = '2')
ax.plot(x , x**3 , label = '3')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('demo plot')
plt.legend(loc = 'upper center')
```

```python
#面向过程
x2 = np.linspace(0,10,100)
plt.plot(x2 , np.sin(x) , label = 'sinx')
plt.plot(x2 , np.cos(x) , label = 'cosx')
plt.xlabel('x')
plt.ylabel('x')
plt.title('demo2 plot')
plt.legend()
```

```python
#练习 利用此函数通过id加一列学院，通过学院进行groupby
def id_to_school(id):
    id = str(id)
    if id.startswith('01',5,7):
        return '机械'   
    if id.startswith('02',5,7):
        return '材料'
    if id.startswith('03',5,7):
        return '能动'
    if id.startswith('04',5,7):
        return '电气'
    if id.startswith('05',5,7):
        return '电信'
    if id.startswith('06',5,7):
        return '航天'
    if id.startswith('09',5,7):
        return '物理'
    if id.startswith('16',5,7):
        return '化工'
    if id.startswith('18',5,7):
        return '公管'
    if id.startswith('19',5,7):
        return '经金'
    if id.startswith('22',5,7):
        return '人居'
    if id.startswith('32',5,7):
        return '新媒体'
    if id.startswith('99',5,7):
        return '非全'
```

```python
def id_to_xz(id):    
    id = str(id)
    if id.startswith('1',4):
        return '学硕'       
    if id.startswith('3',4):
        return '专硕'
    
def id_to_rx(id):    
    id = str(id)
    if id.startswith('19',2):
        return '19级'       
    if id.startswith('20',2):
        return '20级'
    
def id_to_degree(id):    
    id = str(id)
    if id.startswith('2'):
        return '本科'       
    if id.startswith('3'):
        return '硕士'
    if id.startswith('4'):
        return '博士'

grads['degree'] = '0'
grads['入学年份'] = '0'
grads['学硕//专硕'] = '0'
for i in range(len(grads.index)):
        grads.iat[i, 17] = id_to_degree(grads.index[i])
        grads.iat[i, 18] = id_to_rx(grads.index[i])
        grads.iat[i, 19] = id_to_xz(grads.index[i])

grads[['degree','入学年份','学硕//专硕']]
```

```python
grads.to_excel('new.xlsx')
```
