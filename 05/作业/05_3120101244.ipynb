```python
import pandas as pd
import numpy as np
print(pd.Series([2,3,4,5]))
print(pd.Series([2,3,4,5],index=['a','b','c','d']))
print(pd.Series(np.arange(5)))
sr=pd.Series([2,3,4,5],index=['a','b','c','d'])
print(sr[0])
print(sr[0:2])
print(np.abs(sr))
print(sr[sr>3])
sr=pd.Series({'a':1,'b':2})
print(sr['a'])
print('a'in sr,'c'in sr)
for i in sr:
    print(i)
print(sr.index,sr.index[0])
print(sr.values)
sr=pd.Series([2,3,4,5],index=['a','b','c','d'])
print(sr[['a','b']],sr['a':'c'])
sr1=pd.Series(np.arange(20))
sr2=sr1[10:].copy()
print(sr2[10])
#标签与下标
print(sr2.loc[10],sr2.iloc[9])
#数据对齐按标签对齐
sr1=pd.Series([12,23,34],index=['c','a','d'])
sr2=pd.Series([11,20,10,21],index=['d','c','a','b'])
print(sr1+sr2)
#缺失值的处理方法
sr=sr1+sr2
print(sr.isnull())
print(sr[sr.notnull()])
print(sr.dropna())
print(sr.fillna(0))
#DataFrame:二位数据对象
print(pd.DataFrame({'one':[1,2,3],'two':[4,5,6]},index=['a','b','c']))
df=pd.DataFrame({'one':pd.Series([1,2,3],index=['a','b','c']),'two':pd.Series([1,2,3,4],index=['b','a','c','d'])})
print(df)
cd=pd.read_csv('tips.csv')
print(cd)
#index获取行索引 T转置 columns获取列索引 values获取值数组 describe()获取快速统计
print(cd.index)
print(cd.values)
print(cd.columns)
print(cd.T)
#一列要是一个数据类型
#返回每一列的操作
print(cd.describe())
#先选列再选行
print(cd['day'][0])
#loc先行再列
print(cd.loc[0]['day'])
print(cd['day'])
print(cd.loc[0,:])
#DataFrame-数据对齐与缺失数据
print(df.dropna())
df.loc['d','two']=np.nan
df.loc['c','two']=np.nan
print(df.dropna(),df.dropna(how='all'))#全是nan才被删掉
df.loc['d','one']=1
#0为以行为轴 1为以列为轴
print(df.dropna(axis=1))
print(df.mean())
print(df.mean(axis=1))
#排序
print(df.sort_values(by='two'))
#倒序
print(df.sort_values(by='two',ascending=False))
print(df.sort_values(by='a',ascending=False,axis=1))
#nan不参与排序将会放在最后面
print(df.sort_index())
```

```python
import random
import matplotlib.pyplot as plt
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai,color='b',label='上海')
plt.yticks(range(0,40,5))
x_label=["11点{}分".format(i) for i in x]
plt.xticks(range(0,60,5),x_label[::5])
plt.grid(True,linestyle='--',alpha=0.5)
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("温度显示")
y_beijing=[random.uniform(0,3)for i in x]
plt.plot(x,y_beijing,color='r',linestyle='--',label='北京')
plt.ylabel(range(0,40,5))
#显示图例
#plt.legend(loc='lower left')
plt.legend(loc=4)
plt.show()
```
