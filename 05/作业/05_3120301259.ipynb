```python
import time
def show_info():
    print('''
    0
    1
    ''')
    
def write_loginfo(username):
    """
    
    :param username
    """
    with open('log.txt','a') as f:
        string="{}{}\n".format(username,time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(time.time())))
        f.write(string)
        
def read_loginfo():
    """
    
    """
    with open('log.txt','r') as f:
        while True:
            line=f.readline()
            if line=='':
                break
            print(line)
            

username=input('')
while len(username)<2:
    print("2")
    username=input('')
password=input("")
while len(password)<6:
    print('6')
    password=input("")
print('')
write_loginfo(username)
show_info()
num=int(input(''))
while True:
    if num==0:
        print('')
        break
    elif num==1:
        print('')
        read_loginfo()
        show_info()
        num=int(input(''))
    else:
        print("")
        show_info()
        num=int(input(''))
```

```python
def find_answer(question):
    with open('reply.txt','r') as f:
        while True:
            line=f.readline()
            if not line:
                break
            keyword=line.split('|')[0]
            reply=line.split('|')[1]
            if keyword in question:
                return reply
        return ''
question=input('')
while True:
    if question=='bye':
        break
    reply=find_answer(question)
    if not reply:
        question=input("bye:")
    else:
        print(reply)
        question=input("bye:")
print('')
```
