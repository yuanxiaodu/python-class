```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'go', t2, f(t2), 'C6*')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
np.random.seed(19680801)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

```python
def id_to_school(id):
    id = str(id)
    if id.startswith('3120101' or '3120301' or '4120101'):
        return ''
    elif id.startswith('3120103' or '3120303' or '3420999'):
        return ''
    elif id.startswith('3120104' or '3120304'):
        return ''
    elif id.startswith('3120305'):
        return ''
    elif id.startswith('3120105' or '3120305'):
        return ''
    elif id.startswith('3120132' or '3120332' or '3120999'):
        return ''
    elif id.startswith('3120109'):
        return ''
    elif id.startswith('3120118'):
        return ''
    elif id.startswith('3120319'):
        return ''
```
