```python
fig = plt.figure(figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')

for i, category in enumerate(categories):
    plt.scatter('area', 'poptotal', data=midwest.loc[midwest.category==category, :])
```

```python
def encircle(x, y, ax=None, **kw):
                if not ax: ax=plt.gca()
                p = np.c_[x, y]
                hull = ConvexHull(p)
                poly = plt.Polygon(p[hull.vertices, :], **kw)
                ax.add_patch(poly)
```

```python
midwest_encircle_data = midwest.loc[midwest.state=='IN', :]
```

```python
encircle(midwest_encircle_data.area, midwest_encircle_data.poptotal, ec="k", fc="gold", alpha=0.1)
encircle(midwest_encircle_data.area, midwest_encircle_data.poptotal, ec="firebrick", fc="none", linewidth=1.5)
```

```python
import numpy as np
```

```python
a = 1
x = np.arange(-2*np.pi, 2*np.pi, 0.2)
y = np.sin(x) * np.cos(x)
f = np.sin(x) + np.cos(x)
xz = a*(2*np.cos(x) - np.cos(2*x))
yz = a*(2*np.sin(x) - np.sin(2*x))
```

```python
fig = plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)

for i,ax in enumerate(fig.axes):
    ax.plot(x, f, label='Line1 ax%d')
    ax.scatter(x, y, label = u'Markers ax%d', color='r')
    ax.plot(xz, yz, label = u'Line2 ax%d')
    ax.grid(True)
    ax.set_xlabel(u'Argument')
    ax.set_ylabel(u'Function')
    ax.set_title(u'Multiple charts')
    ax.legend(loc='best', frameon=False)

plt.tight_layout()

plt.show()
```
