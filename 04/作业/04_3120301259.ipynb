```python
import face_recognition
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
image=face_recognition.load_image_file("akb48.jpeg")
face_locations=face_recognition.face_locations(image)
im=np.array(Image.open('akb48.jpeg'),dtype=np.uint8)
fig,ax=plt.subplots(1,figsize=(8,8))
ax.imshow(im)
for face_loc in face_locations:
    rect=patches.Rectangle((face_loc[1], face_loc[0]), face_loc[3] - face_loc[1], face_loc[2] - face_loc[0], linewidth=2,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import array
from numpy.random import normal
def getData():
    heights=[]
    weights=[]
    books=[]
    N=10000
    for i in range(N):
        while True:
            height=normal(172,6)
            if 0<height:
                break
        while True:
            weight=(height-80)*0.7+normal(0,1)
            if 0<weight:
                break
        while True:
            number=normal(20,5)
            if 0<=number and number<=50:
                book='E' if number<10 else('D'if number<15 else('C'if number<20 else('B' if number<25 else 'A')))
                break
        heights.append(height)
        weights.append(weight)
        books.append(book)
    return array(heights),array(weights),array(books)
heights,weights,books=getData()
def drawBar(books):
    x=['A','B','C','D','E']
    bookgroup={}
    for book in books:
        bookgroup[book]=bookgroup.get(book,0)+1
    plt.bar(range(5),[bookgroup.get(x1,0) for x1 in x],align='center')
    plt.xticks(range(5),x)
    plt.xlabel("Types of students")
    plt.ylabel('Frequency')
    plt.title("Numbers of books students read")
    plt.show()       
drawBar(books)
```

```python
#对上面books画饼状图
def drawPie(books):
    labels={'A','B','C','D','E'}
    bookgroup={}
    for book in books:
        bookgroup[book]=bookgroup.get(book,0)+1
    plt.pie([bookgroup.get(label,0) for label in labels],labels=labels,autopct='%.1f%%')
    plt.title("Number of books students read")
    plt.show()
drawPie(books)
```

```python
#对上面的heights画直方图
def drawHist(heights):
    plt.hist(heights,100)
    plt.xlabel('Heights')
    plt.ylabel('Frequency')
    plt.title('Heights of Students')
    plt.show()
drawHist(heights)
```

```python
#绘制累加曲线
def drawCumulativaHist(heights):
    plt.hist(heights,20,histtype='step',cumulative=True)
    plt.xlabel('Heights')
    plt.ylabel('Frequency')
    plt.title('Heights of students')
    plt.show()
drawCumulativaHist(heights)
```

```python
#绘制散点图
def drawScatter(heights,weights):
    plt.scatter(heights,weights)
    plt.xlabel('Heights')
    plt.ylabel('Weights')
    plt.title('Heights & Weights of students')
    plt.show()
drawScatter(heights,weights)
```

```python
#绘制箱型图
def drawbox(heights):
    plt.boxplot([heights],labels=['Heights'])
    plt.title('Heights of students')
    plt.show()
drawbox(heights)
```
