```python
import pynlpir
def cut(data):   #分词
    pynlpir.open()
    segments=pynlpir.segment(data,pos_english=False)
    pynlpir.close()
    return segments
r2020=reports.loc[2020,'txt']
cut(r2020)
```

```python
import collections
def cp(data):  #单词计算词次数
    pynlpir.open()
    segments=pynlpir.segment(data,pos_english=False)
    pynlpir.close()
    df=pd.DataFrame(segments)
    df.columns=['词','词性']
    df=df[df.词性==df.词性]
    df=df[df.词性!='标点符号']
    df=df[df.词性!='介词']
    df=df[df.词性!='助词']
    df=df[df.词性!='数词']
    df=df[df.词性!='连词']
    df['词']=df.词.apply(lambda x: x.replace('\n',''))
    x=collections.Counter(df.词)
    词数=pd.DataFrame(x,index=['count']).T
    词数['词频']=词数['count']/词数['count'].sum()*10000
    return 词数[['词频']].copy()
df2020=cp(r2020)
df2020
```

```python
df=pd.DataFrame()
for y in reports.index[:10]:
    df0=cp(reports.loc[y,'txt'])
    df0.columns=[y]
    df=pd.concat([df,df0],axis=1,sort=False)
df
```

```python
# pynlpir版
import pynlpir
import re
import pandas as pd
keys=['不确定','波动','不清晰','不明朗','难以预测','无法估计','政策','制度','改革','整改','监管','赤字']
df=pd.DataFrame(columns=['Y','Uncertainty'])
with open('历年政府工作报告.txt','r',encoding='gbk',errors='ignore') as file:
    data=file.read()
x=data.split('年政府工作报告')
for i in range(1,10):
    data=x[i].replace('\n','')
    y=re.findall('([0-9]+)',x[i])[0]
    pynlpir.open()
    cut=pynlpir.segment(data,pos_english=False)
    pynlpir.close()
    lcut=[]
    for j in cut:
        if j in cut:
            if j[1] not in ['标点符号','介词','助词','数词','连词']:
                lcut.append(j[0])
    p=0
    for k in keys:
        p+= lcut.count(k)/len(lcut)*100
    df.loc[df.shape[0]+1]={'Y':y,'Uncertainty':p}
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets as skds
def mydistance(x,y,q=2):
    return(np.sum(np.abs(x-y)**q))**(1/q)
def myKmens(k,data,maxloop,scatter=False):
    datalen=len(data)
    centers=np.array([data[i] for i in np.random.randint(0,datalen,size=k)])
    for _ in range(maxloop):
        target_1=np.full(shape=datalen,fill_value=-1)
        for i in range(datalen):
            distinces=[]
            for j in range(k):
                distinces.append(mydistance(data[i],centers[j]))
            target_1[i]=np.argmin(distinces)
        if scatter:
            fig=plt.figure(figsize=(6,4))
            for j in range(k):
                cluster_j=data[target_1==j]
                plt.scatter(cluster_j[:,0],cluster_j[:,1])
                plt.scatter(centers[j][0],centers[j][1],s=350,marker='+',c='blue')
        centers_new=np.array([np.mean(data[target_1==j],axis=0) for j in range(k)])
    if not(centers_new==centers).all():
        print('新旧中心有变化，继续迭代更新...')
        centers=centers_new.copy()
    else:
        return target_1
    return target_1
data,target=skds.make_blobs(n_features=2,centers=3,cluster_std=1)
target_1=myKmens(3,data,20,True)
print(target,target_1)
```

```python
from sklearn.cluster import KMeans
mykm=KMeans(n_clusters=3)
mykm.fit(data)
target_sklearn=mykm.predict(data)
plt.scatter(data[:,0],data[:,1],c=target_sklearn)
```

```python
def mySSE(cluster):
    mysum=0
    center=np.mean(cluster)
    for a in cluster:
        mysum+=np.sum(np.square(a-center))
    return mysum
```

```python
target_1=myKmens(3,data,20)
sse=0
for j in range(3):
    clusterJ=data[target_1==j]
    sse+=mySSE(clusterJ)
print('误差平方和：',sse)
```

```python
from sklearn.cluster import KMeans
ar=[]
for k in range(2,50):
    sse=0
    mykm=KMeans(n_clusters=k)
    mykm.fit(data)
    y_predict_sklearn=mykm.predict(data)
    for j in range(k):
        sse+=mySSE(data[y_predict_sklearn==j])
    ar.append(sse)
plt.plot(ar)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('成绩.xlsx')
df.head()
```

```python
mymap={'YES':0,'NO':1}
df.married=df.married.map(mymap)
df.car=df.car.map(mymap)
df.save_act=df.save_Act.map(mymap)
mymap={'FEMALE':0,'MALE':1}
df.sex=df.sex.map(mymap)
df.head()
df.drop()
```

```python
from sklearn.feature_extraction import DictVectorizer
dv=DictVectorizer(sparse=False)
data=df[['region']].to_dict(orient='records')
dvdata=dv.fit_transform(data)
df_Drug=pd.DataFrame(dvdata,columns=dv.feature_names_)
df_new=pd.concat([df.drop(['region','id'],axis=1),df_Drug],axis=1)
df_new.head()
```

```python
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
df_scaler=scaler.fit_transform(df_new)
df_scaler=pd.DataFrame(df_scaler,columns=df_new.columns)
df_scaler.head()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=7)
km.fit(df_scaler)
y_predict=km.predict(df_scaler)
drugdf_result=pd.concat([df_scaler,pd.DataFrame(y_predict,columns=['RESULT'])],axis=1)
drugdf_result.head()
```
