```python
from lxml import etree

text = '''
<div>
    <ul>
         <li class="item-0"><a href="link1.html">first item</a></li>
         <li class="item-1"><a href="link2.html">second item</a></li>
         <li class="item-inactive"><a href="link3.html">third item</a></li>
         <li class="item-1"><a href="link4.html">fourth item</a></li>
         <li class="item-0"><a href="link5.html">fifth item</a>
     </ul>
 </div>
'''
# 将字符串解析为html文档,如果不规范，可以补全
html = etree.HTML(text)
print(html)
# 按字符串序列化html
string = etree.tostring(html).decode("utf-8")
print(string)

# 也可以读取html文件
html = etree.parse("hello.html")
print(html)
string = etree.tostring(html).decode("utf-8")
print(string)
```

```python
from lxml import etree
html = etree.parse('hello.html')
# 获取所有li标签：通过for循环取出
result = html.xpath("//li")
for i in result:
    print(etree.tostring(i).decode("utf-8"))

```

```python
# 获取所有li元素下的所有class属性的值：
result = html.xpath("//li/@class")
print(result)
```

```python
# 获取li标签下href为www.baidu.com的a标签：
result = html.xpath('//li/a[@href="www.baidu.com"]')
print(result)
```

```python
# 获取li标签下所有span标签：单斜杠/是匹配直接子节点   //是所有子节点
result = html.xpath('//li//span')
print(result)
```

```python
# 获取li标签下的a标签里的所有class：
result = html.xpath('//li/a//@class')
print(result)

# 获取最后一个li的a的href属性对应的值：
result = html.xpath('//li[last()]/a/@href')
print(result)

# 获取倒数第二个li元素的内容：
result = html.xpath('//li[last()-1]/a')
print(result)
print(result[0].text)

# 获取倒数第二个li元素的内容的第二种方式：
result = html.xpath('//li[last()-1]/a/text()')
print(result)
```

```python
response = requests.get(url=url, headers=headers)
text = response.content.decode("utf-8")
html = etree.HTML(text)
print(html)
```

```python
ul = html.xpath('//ul[@class="carlist clearfix js-top"]')[0]  # xpath 语法最好自己写多观察网页结构  返回list 取出第一个 虽然只有一个
# print(type(ul))
li = ul.xpath('./li')  # 当前节点下的li标签
detail_urls = []
for i in li:
    detail_url = i.xpath('./a/@href')  # 返回未补全的url,仍然是一个列表
    detail_url = "https://www.guazi.com" + detail_url[0]
    detail_urls.append(detail_url)
    print(detail_url)
def get_image(url):
    """
    获取挂牌时间
    由于挂牌时间使用图片形式给出，因此输入图片url 获取图片
    并保存到本地，使用ocr识别图片文字 并保存
     :param url: 图片的url
    """
    time = []
    # print(url)
    response = requests.get(url)
    with open("c.jpg", "wb") as f:
        f.write(response.content)  # response.content 是二进制文件
    text = pytesseract.image_to_string(Image.open("c.jpg"), lang="chi_sim")
    # print(text)
    time.append(text.replace("\n", ""))  # 这里注意转义字符的问题，发现每个text之后有两个回车，因此是\n ，而不是r“\n” 这代表\n字符
    return time
```

```python
datalist = []  # 保存所有车的信息 放在外面
for d_url in detail_urls:
#     sleep(10)
    response = requests.get(url=d_url, headers=headers)
    text = response.content.decode("utf-8")
    html = etree.HTML(text)
#     print(html)
    data = []  # 保存一辆车的所有信息 这个list一定要放在for循环里面，每次重置为空列表，不然最后datalist只有一个list元素
    # 获取标题
    title = html.xpath("//div[@class='product-textbox']/h1/text()")[0]
    if title:  # 空列表为false
        data.append(title.replace(r'\r\n', "").strip())
        # print(title)

        # 获取挂牌时间
        img_url = html.xpath("//div[@class='product-textbox']/ul/li[@class='one']/span/img/@src")[
            0]  # [0]返回列表第一个元素 不然变成二维列表  ,这里每个img_url都是单独的一个url
        # data.append(img_url)  # 把图片url放入列表
        time = get_image(img_url)
        # print(time)  # 打印上牌时间
        data.append(time[0])
        # print(img_url)

        # 获取里程数
        mileage = html.xpath("//div[@class='product-textbox']/ul/li[@class='two']/span/text()")[0]
        data.append(mileage)
        # print(mileage)

        # 获取排量
        displacement = html.xpath("//div[@class='product-textbox']/ul/li[@class='three']/span/text()")[0]
        data.append(displacement)
        # print(displacement)

        # 变速方式  自动 or 手动
        trans = html.xpath("//div[@class='product-textbox']/ul/li[@class='last']/span/text()")[0]
        data.append(trans)
        # print(trans)

        # 获取价格
        price = html.xpath(
            "//div[@class='product-textbox']/div[@class='pricebox js-disprice']/div[@class='price-main']/span/text()")[
            0]
        data.append(price)
        # print(price)

        datalist.append(data)  # 每辆车的信息是一个元素（list）
print(datalist)
```

```python
sql = '''
create table guazi
(
id integer primary key autoincrement,
titles text,
time text,
mileages text,
displacements text,
transmission text,
prices text
)
'''  # 创建数据表
conn = sqlite3.connect(dbpath)
cursor = conn.cursor()
cursor.execute(sql)
conn.commit()
cursor.close()
conn.close()
print('Table created successfully')
```
