```python
import xlrd
import numpy as np
from pyecharts.charts import Bar
from pyecharts.charts import Pie, Grid
from pyecharts import options as opts

# 导入Excel 文件
data = xlrd.open_workbook(fpath)
# 载入表格
table = data.sheets()[0]
tables = []

for rows in range(1, table.nrows ):
    dict_ = {'name':'','IDCard':'','age':'','area':''}
    dict_["name"] = table.cell_value(rows, 0)
    dict_["IDCard"] = table.cell_value(rows, 1)
    dict_["age"] = table.cell_value(rows, 2)
    dict_["area"] = table.cell_value(rows, 3)
                # 过滤非个体公民的信息
    if dict_["age"] != '0':
        tables.append(dict_)

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
age_20 = 0
age_21_25 = 0
age_26_30 = 0
age_31_35 = 0
age_36_40 = 0
age_41_45 = 0
age_46_50 = 0
age_51_55 = 0
age_56_60 = 0
age_61_65 = 0
age_66_70 = 0
age_71 = 0

for i in tables:
    num_age = i["age"]
    num_age = int(num_age)
    if num_age in range(0, 20):
        age_20 = age_20 + 1
    elif num_age in range(21, 25):
        age_21_25 = age_21_25 + 1
    elif num_age in range(26, 30):
        age_26_30 = age_26_30 + 1
    elif num_age in range(31, 35):
        age_31_35 = age_31_35 + 1
    elif num_age in range(36, 40):
        age_36_40 = age_36_40 + 1
    elif num_age in range(41, 45):
        age_41_45 = age_41_45 + 1
    elif num_age in range(46, 50):
        age_46_50 = age_46_50 + 1
    elif num_age in range(51, 55):
        age_51_55 = age_51_55 + 1
    elif num_age in range(56, 60):
        age_56_60 = age_56_60 + 1
    elif num_age in range(61, 65):
        age_61_65 = age_61_65 + 1
    elif num_age in range(66, 70):
        age_66_70 = age_66_70 + 1
    elif num_age in range(71, 150):
        age_71 = age_71 + 1

# 构建两个元组用以后期建表方便
bar_x_axis_data = ("20及以下", "21-25", "26-30", "31-35", "36-40", "41-45", "46-50", "51-55", "56-60", "61-65", "66-70","70以上")
bar_y_axis_data = (age_20, age_21_25, age_26_30,
                   age_31_35, age_36_40, age_41_45,
                   age_46_50, age_51_55, age_56_60,
                   age_61_65, age_66_70, age_71)
#柱状图
c = (
    Bar()
        .add_xaxis(bar_x_axis_data)
        .add_yaxis("失信人各年龄段数量", bar_y_axis_data, color="#af00ff")
        .set_global_opts(title_opts=opts.TitleOpts(title="数量"))
#         .render_notebook()
        .render("./失信人各年龄段数量柱状图.html")
    
)

```
