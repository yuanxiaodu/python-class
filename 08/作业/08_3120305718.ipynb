```python
##构造空的list存储数据
data = [[]for i in range(5)]
data
```

```python
index = 0
elem = ["OCCTIME","MERCNAME","TRANAMT","TRANNAME","CARDBAL"]
for i in elem:
    record = soup.find_all("td",field=i)  # 把五个属性分别放到data[[elem]]  里
    for j in record:
        data[index].append(j.text)
    index += 1
```

```python
for i in range(len(data)):
    print(len(data[i]))
```

```python
import pandas as pd
import numpy as np
array = np.array(data)
array= array.T
datas = pd.DataFrame(columns=elem,data=array)

datas.reset_index(drop = True)
datas.head()
datas.to_csv("record.csv",encoding='utf-8',index=False)
```

```python
def delblankline(infile, outfile):
    infopen = open(infile, 'r',encoding="utf-8")
    outfopen = open(outfile, 'w',encoding="utf-8")
    db = infopen.read()
#     outfopen.write(db.replace('\n',' '))
    outfopen.write(db.replace('创新港水控','洗澡'))
    outfopen.write(db.replace('创新港电控收费',''))
    infopen.close()
    outfopen.close()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 生成词云
def create_word_cloud(txt):
    frequencies = {}
    for line in open(txt,encoding='utf-8'):
        arr = line.split(" ")
        frequencies[arr[0]] = float(arr[1])
    wc = WordCloud(
        font_path="C:/Windows/Fonts/SimHei.ttf",
        max_words=100,
        width=2000,
        height=1200,
    )
    word_cloud = wc.generate_from_frequencies(frequencies)
    # 写词云图片
    word_cloud.to_file("wordcloud2.jpg")
    # 显示词云文件
    plt.imshow(word_cloud)
    plt.axis("off")
    plt.show()
```

```python
# 根据词频生成词云
txt = 'ciyunnew1.txt'
create_word_cloud(txt)
```
