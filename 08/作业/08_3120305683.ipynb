```python
import pandas as pd
from snownlp import SnowNLP

# 显示所有列
pd.set_option('display.max_columns', None)
# 显示所有行
pd.set_option('display.max_rows', None)
# 设置value的显示长度为100，默认为50
pd.set_option('max_colwidth', 100)
    
tweets = pd.read_csv(r'C:/Users/Administrator/Desktop/newcode/1720042973.csv', usecols=['正文'])
tweets['snownlp分词'] = tweets.apply(lambda x: ';'.join(SnowNLP(x['正文']).words), axis=1)
tweets['emotional_grade'] = tweets.apply(lambda x: SnowNLP(x['正文']).sentiments, axis=1)
tweets
```

```python
import seaborn as sns
import pandas as pd
sns.set_theme(style="darkgrid")
# Samuel Norman Seaborn is a fictional character portrayed 
# by Rob Lowe on the television serial drama The West Wing. 
sns.histplot(data=tweets, x="emotional_grade",multiple="stack")
```
