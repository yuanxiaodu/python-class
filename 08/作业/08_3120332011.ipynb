### 年月份分析

```python
data.head()
```

```python
data['year'] = data['pub_time'].str.split('-').str[0]
data['month'] = data['pub_time'].str.split('-').str[1]
```

```python
data.head()
```

```python
year = data.groupby('year')['year'].count()
year
```

```python
from pyecharts import Line

attr = list(year.index)
v = list(year)
line = Line("电影年份分布情况")
line.add("", attr, v, mark_point=["average"])
line
```

```python
month = data.groupby('month')['month'].count()
month
```

```python
from pyecharts import Bar

attr = list(month.index)
v = list(month)
bar = Bar("")
bar.add("", attr, v)
bar
```

### 地区

```python
data.head()
```

```python
def get_country(i):
    country = i.split('(')
    if len(country) == 1:
        return '中国'
    else:
        country_1 = country[1].strip(')')
        if country_1 == '中国香港':
            return '中国'
        elif country_1 == '法国戛纳':
            return '法国'
        else:
            return country_1
```

```python
data['country'] = data['pub_time'].map(get_country)
data.head()
```

```python
country = data.groupby('country')['country'].count()
country
```

```python
from pyecharts import Pie

attr = list(country.index)
v1 = list(country)
pie = Pie("", title_pos='center')
pie.add(
    "",
    attr,
    v1,
    radius=[40, 75],
    label_text_color=None,
    is_label_show=True,
    legend_orient="vertical",
    legend_pos="left",
)
pie
```

### 明星上榜

```python
data.head()
```

```python
str = ''
for i in range(100):
    str = str + data.iloc[i,1] + ','
str
```

```python
author = str.split(',')
author
```

```python
from collections import Counter
c = Counter(author)
c
```

```python
count = c.most_common(6)
count
```

```python
attr = []
v = []
for i in count:
    attr.append(i[0])
    v.append(i[1])
```

```python
attr
```

```python
v
```

```python
from pyecharts import Bar
bar = Bar("")
bar.add("", attr, v)
bar
```
