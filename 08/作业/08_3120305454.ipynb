```python
url = "https://s.weibo.com/top/summary?cate=realtimehot"

# 请求头
headers={
    'Host': 's.weibo.com',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'zh-CN,zh;q=0.9',
    'Connection': 'keep-alive',
    'Referer': 'https://weibo.com/',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.97 Safari/537.36'
}

# 发起请求
r = requests.get(url,headers=headers)
print(r.status_code)


# 解析xml
html_xpath = etree.HTML(r.text)
data = html_xpath.xpath('//*[@id="pl_top_realtimehot"]/table/tbody/tr/td[2]')
num = -1


# 组织存储路径
time_path = time.strftime('%Y{y}%m{m}%d{d}',time.localtime()).format(y='年', m='月', d='日')
time_name = time.strftime('%Y{y}%m{m}%d{d}%H{h}',time.localtime()).format(y='年', m='月', d='日',h='点')
year_path = time.strftime('%Y{y}',time.localtime()).format(y='年')
month_path = time.strftime('%m{m}',time.localtime()).format(m='月')
day_month = time.strftime('%d{d}',time.localtime()).format(d='日')

all_path = "./data/" + year_path + '/'+ month_path + '/' + day_month
if not os.path.exists(all_path):
    # 创建多层路径
    os.makedirs(all_path)


# 形成文件路径
root = all_path  + "/"
path = root + time_name + '.md'

print(path)
# 文件头部信息
with open(path,'a') as f:
    f.write('{} {}\n\n'.format('# ',time_name+'数据'))
f.close()

# 打印查看
for tr in (data):
    title = tr.xpath('./a/text()')
    hot_score = tr.xpath('./span/text()')
    
    num += 1

    # 过滤第 0 条
    if num == 0:
        pass
    else:
        with open(path,'a') as f:

            f.write('{} {}、{}\n\n'.format('###',num,title[0]))
            f.write('{} {}\n\n'.format('微博当时热度为：',hot_score[0]))
         
        f.close()

        print(num,title[0],'微博此时的热度为：',hot_score[0])
```

## 抽取整理数据

<div align=center>
<img src="./images/file.png" width="600" height="600" />
</div>

```python
# 对文件进行抽取
def par_data(path,time):
    timestamp=[]
    hot_num=[]
    title = []
    index = []
    with open(path, "r") as f:
        for line in f:
            if line.isspace() == True or '点数据' in line:
                continue
            if '###' in line:
                line =line.replace('#','')
                line =line.replace('\n','')
                res = line.split('、')
                day_index = res[0]
                the_title = res[1]
                index.append(day_index)
                title.append(the_title)
                timestamp.append(time)
                
            if '微博当时热度为' in line:
                hot = re.sub("\D", "", line)
                hot_num.append(hot)                
    res = pd.DataFrame([timestamp,index,hot_num,title]).T
    res.columns=['time','hot_index','hot_num','news_title']
    return res


# 迭代文件    
def iter_file(year_path,year):
    total=pd.DataFrame()
    year_files = os.listdir(year_path)
    year_files = sorted(year_files)
    for month_name in year_files:
        month_path = os.path.join(year_path,month_name)
        if '.DS_Store' in month_path:
                continue
        month_files = os.listdir(month_path)
        month_files = sorted(month_files)
    
        for day_name in month_files:
            day_path = os.path.join(month_path,day_name)
            if '.DS_Store' in day_path:
                continue
            day_files = os.listdir(day_path)
            
            for day_time in day_files:
                day_time_path = os.path.join(day_path,day_time)
                if '.DS_Store' in day_time_path:
                    continue
#                 抽取
                day_df = par_data(day_time_path,day_time.replace('.md',''))
                if day_df.shape[0]<20:
                    print(day_time_path)
#                 拼接
                total =  pd.concat([total,day_df],ignore_index=True)
    return total
```

### 整理数据主要有：热搜榜时间，热搜度，热搜标题，热搜排名。

# 对所有标题进行命名实体识别

```python
def ner_by_LTP(text):
    from ltp import LTP
    ltp = LTP()
    cut_word = []
    cut_word_type=[]
    for t in text:
        seg, hidden = ltp.seg([t])
        ner = ltp.ner(hidden)
        if len(ner[0])==0:
            continue
        for ner_t in ner[0]:
            tag, start, end = ner_t
            cut_word.append("".join(seg[0][start:end + 1]))
            cut_word_type.append(tag)
    return cut_word,cut_word_type
```

## 2020 最爆热搜

### 每季度最热

```python
seasons = ['2020-03-01 11:00:00','2020-06-01 11:00:00','2020-09-01 11:00:00','2021-01-01 11:00:00']
start = '2020-01-01 11:00:00'
for index,season in enumerate(seasons):
    end = season
    t = total[(total['time'] >= start) & (total['time'] <= end)]
    hot_t = t.sort_values(by=['hot_num'],ascending=False).head(20)[['news_title','hot_num']]
    fig = plt.figure(figsize=(20,10))
    plt.yticks(fontsize=25)

    plt.title(str(index+1)+'季度热搜',fontdict = myfontdict)
    g = sns.barplot(x=hot_t['hot_num'],y=hot_t['news_title'])
    plt.xlabel('x titile',fontsize=13)
    g.set_xlabel('热度（千万）',fontsize=20)
    g.set_ylabel('热搜标题',fontsize=30)
    start  = end
```
