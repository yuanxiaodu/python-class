### 性别分析

```python
sex = data.groupby('Sex')['Sex'].count()
sex
```

```python
from pyecharts import Pie

attr = ['外星人','男性','女性']

v = list(sex)
pie = Pie("微信好友性别分布",  width=900)
pie.add(
    "",
    attr,
    v,
    center=[50, 50],
    is_random=True,
    radius=[30, 75],
    rosetype="radius",
)
pie
```

### 个性签名词云图

```python
data.head()
```

```python
new_data2 = data[data['Signature'].notnull()]
new_data2.head()
```

```python
str_data = ''
for i in range(new_data2.shape[0]):
    str_data = str_data + new_data2.iloc[i,4]
str_data
```

```python
import re
str_data = re.sub('span', '',str_data,re.S)
str_data = re.sub('class', '',str_data,re.S)
str_data = re.sub('emoji', '',str_data,re.S)
str_data
```

```python
import jieba.analyse
tags = jieba.analyse.extract_tags(str_data, topK=50, withWeight=True)
```

```python
label = []
attr = []
for item in tags:
    print(item[0]+'\t'+str(int(item[1]*1000)))
    label.append(item[0])
    attr.append(int(item[1]*1000))
```

```python
from pyecharts import WordCloud

wordcloud = WordCloud(width=800, height=620)
wordcloud.add("", label[4:], attr[4:], word_size_range=[20, 100])
wordcloud
```
