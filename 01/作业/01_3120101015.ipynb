```python
#用python实现基于Givens、Householder的QR分解

import numpy as np

def givens_rotation(A):          #创建函数
    """Givens变换"""
    (m, n) = np.shape(A)
    Q = np.identity(m)       #获得单位矩阵
    R = np.copy(A)
    (rows, cols) = np.tril_indices(m, -1, n)     
    for (row, col) in zip(rows, cols):
        if R[row, col] != 0:  # R[row, col]=0则c=1,s=0,R、Q不变
            m_ = np.hypot(R[col, col], R[row, col])  # d
            n = R[col, col]/m_
            s = -R[row, col]/m_
            G = np.identity(m)
            G[[col, row], [col, row]] = n
            G[row, col] = s
            G[col, row] = -s
            R = np.dot(G, R)  # R=G(n-1,n)*...*G(2n)*...*G(23,1n)*...*G(12)*A
            Q = np.dot(Q, G.T)  # Q=G(n-1,n).T*...*G(2n).T*...*G(23,1n).T*...*G(12).T
    return (Q, R)

def householder_reflection(A):
    """Householder变换"""
    (r, c) = np.shape(A)
    Q = np.identity(r)
    R = np.copy(A)
    for cnt in range(r - 1):
        x = R[cnt:, cnt]
        e = np.zeros_like(x)
        e[0] = np.linalg.norm(x)
        u = x - e
        v = u / np.linalg.norm(u)
        Q_cnt = np.identity(r)
        Q_cnt[cnt:, cnt:] -= 2.0 * np.outer(v, v)
        R = np.dot(Q_cnt, R)  # R=H(n-1)*...*H(2)*H(1)*A
        Q = np.dot(Q, Q_cnt)  # Q=H(n-1)*...*H(2)*H(1)  H为自逆矩阵
    return (Q, R)

np.set_printoptions(precision=3, suppress=True)
A = np.array([[3,5,5],[0,3,4],[4,0,5]],dtype=float)

(Q, R) = givens_rotation(A)
print('Q=',Q,'\n')
print('R=',R,'\n')
print('A=',np.dot(Q,R),'\n')

(Q, R) = householder_reflection(A)
print('Q=',Q,'\n')
print('R=',R,'\n')
print('A=',np.dot(Q,R),'\n')
```

```python
import math
doc = open("龙贝格积分法的输出结果.txt",'w+')

def func1(x):
     return 1 / (1+x)

def func2(x):
    return (math.log(1+x)) / (1+x**2)

def func3(x):
    if x != 0:
        fx3 = (math.log(1 + x)) / x
    else:
        fx3 = 1/(1+x)
    return fx3

def func4(x):
    if x == 0:
        fx4 = math.cos(x)
    else:
        a = (math.pi)/2
        fx4 = (math.sin(a*x))/(a*x)
    return fx4

def Romberg(a,b,n):
    for t in [func1,func2,func3,func4]:
        T_list = []
        S_list = []
        C_list = []
        R_list = []
        y = (t(a) +t(b))/2
        T = (b-a)*y
        T_list.append(T)
        for i in range(1, n):
            s = 0
            for j in range(1, 2**(i-1)+1):
                r = t(a + (2*j-1)*(b-a) / 2**i)
                s += r
            T = T_list[i-1]/2 + ((b-a)*s) / (2**i)
            T_list.append(T)
        for i in range(1,n-1):
            S = T_list[i] + (T_list[i] - T_list[i-1]) / (4**1 - 1)
            S_list.append(S)
        for i in range(1,n-2):
            C = S_list[i] + (S_list[i] - S_list[i-1]) / (4**2 - 1)
            C_list.append(C)
        R = C_list[n-4] + (C_list[n-4] - C_list[n-5]) / (4 ** 3 - 1)
        print("函数的积分结果为：%.7f" % R)
        print("函数的积分结果为：%.7f" % R,file = doc)
        #  上面是直接输出最终的结果，下面可以输出k=0，1，2，，，的R值
        for i in range(1, n-3):
            R = C_list[i] + (C_list[i] - C_list[i-1]) / (4**3 - 1)
            R_list.append(R)
        print("k=0,1,2,,,时的各个值为：" , R_list)
        print("k=0,1,2,,,时的各个值为：", R_list,file = doc)

if __name__ == '__main__':
    # a = 0，b = 1，n = 8
    Romberg(0,1,8)
    doc.close()
```

```python
# 异常处理
import sys
def fibonacci(n):
    a,b,counter = 0,1,0
    while True:
        if counter>n:
            return
        yield a
        a,b = b,a+b
        counter += 1
f = fibonacci(10)

while True:
    try:
        print (next(f), end=" ")
    except StopIteration:
        sys.exit()
```
