```python
class Person:
    def __init__(self,n,a):
        self.__name=n
        self.__age=a
        pass
    def __str__(self):          #获取函数
        return '{}的年龄是{}'.format(self.__name,self.__age)
    def getInfo(self):          #获取年龄
        return self.__age
    def getName(self):         #获取姓名
        return self.__name
    def setAge(self,age):        #修改年龄
        if age>0 and age<120:
            self.__age=age
        else:
            print('数据不合格')
        pass
# 2  创建一个类，并定义两个私有化属性，提供一个获取属性的方法

class Student:
    def __init__(self):
        self.__name='张三'
        self.__score=90
        pass
    @property
    def name(self):
        return self.__name
    @name.setter
    def name(self,name):
        self.__name=name
        pass
    def __str__(self):
        return ''
    def __call__(self, *args, **kwargs):
        print(self.__name+'的得分是'+str(self.__score))
    pass
xw=Student()
print(xw)
xw()   #将实例对象以函数的形式去调用
```

```python
import types
def run(self):
    print('小猫跑得快')
class Animal:
    pass
@classmethod
def info(cls):
    print('ok')
Animal.color='黑色'  #绑定类属性
Animal.info=info    #绑定类方法
cat=Animal()
cat.run=types.MethodType(run,cat)  #动态绑定
cat.run()
print(cat.color)
Animal.info()
```

```python
def jiecheng(n):
    i=1
    m=1
    while i<=n:
        m*=i
        i+=1
        pass
    return m
print(jiecheng(5))# 循环的方式
def diguiJc(n):
    if n==1:
        return 1
    else:
        return n*diguiJc(n-1)
    pass
print('递归函数的结果是{}'.format( diguiJc(5)) )
```
